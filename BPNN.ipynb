```python
import numpy as np
```

```python
x = np.array(([2,9],[1,5],[3,6]),dtype = float)
y = np.array(([98],[89],[83]),dtype = float)
x = x/np.amax(x,axis=0)
y = y/100
print(x)
print(y)
```

```python
class Neural_networks():
    def __init__(self):
        self.inputlayer = 2
        self.hiddenlayer = 3
        self.outputlayer = 1
        
        np.random.seed(0)

        self.w1 = np.random.randn(self.inputlayer,self.hiddenlayer)
        self.w2 = np.random.randn(self.hiddenlayer,self.outputlayer)
        print("w1:",self.w1)
        print("w2:",self.w2)
        
    def sigmoid(self,s):
        return 1/(1+np.exp(-s))
    
    def sigmoidprime(self,s):
        return s*(1-s)
        
    def forward(self,x):
        self.z = np.dot(x,self.w1)
        self.z2 = self.sigmoid(self.z)
        self.z3 = np.dot(self.z2,self.w2)
        
        O = self.sigmoid(self.z3)
        return O
    
    def backward(self,x,y,O):
        self.O_error = y-O
        self.O_delta = self.O_error*self.sigmoidprime(O)
        self.z2_error = self.O_delta.dot(self.w2.T)
        self.z2_delta = self.z2_error*self.sigmoidprime(self.z2)
        self.w1 += x.T.dot(self.z2_delta)
        self.w2 += self.z2.T.dot(self.O_delta)
        
    def train(self,x,y):
        O = self.forward(x)
        self.backward(x,y,O)
        
```

```python
NN = Neural_networks()
for i in range(2000):
    print("INPUT:\n" + str(x))
    print("ACTUAL OP:\n" + str(y))
    print("PREDICTED OP:\n" + str(NN.forward(x)))
    print("LOSS:\n" + str(np.mean(np.square(y-NN.forward(x)))))
    print("\n")
    NN.train(x,y)
```
