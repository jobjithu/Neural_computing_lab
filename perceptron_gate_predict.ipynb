```python

w1, w2, b = 0,0,0 

def step_fun(x):
    return 1 if x >= 0 else 0

def perceptron(inputs, desired_outputs, learning_rate, epochs):
    global w1,w2,b
    for epoch in range(epochs):
        print('\nepoch:',epoch)
        total_error = 0
        for i in range(len(inputs)):
            A, B = inputs[i]
            target_output = desired_outputs[i]
            print(f"For input({A},{B})")
            output = step_fun(w1 * A + w2 * B + b)
            print('predicted_output:',output)
            error = target_output - output
            print('Error:',error)
            w1 += learning_rate * error * A
            print('w1:',w1)
            w2 += learning_rate * error * B
            print('w2:',w2)
            b += learning_rate * error
            print('b:',b)
            total_error += abs(error)
            print('total_error:',total_error)
        if total_error == 0:
            break
            
            
def score(desired_outputs,inputs):
    predicted = []
    for i in range(len(inputs)):
        A,B = inputs[i]
        output = step_fun(w1 * A + w2 * B + b)
        predicted.append(output)
    score =0
    for i in range(4):
        if(desired_outputs[i]==predicted[i]):
            score +=1
    accuracy = (score/4)*100   
    return accuracy 


inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
desired_outputs = [1,1,1,0]
learning_rate = 0.1
epochs = 10

perceptron(inputs, desired_outputs, learning_rate, epochs)

for i in range(len(inputs)):
    A, B = inputs[i]
    output = step_fun(w1 * A + w2 * B + b)
    print(f"Input: ({A}, {B})  Output: {output}")
    
Accuracy = score(desired_outputs,inputs) 
print(f"Accuracy of the model:{Accuracy}%")
    
```
